```python
map_inscountry_data = {
    'State': [], 'Year': [], 'Quater': [],
    'District': [], 'Metric': [], 'Latitude': [], 'Longitude': []
}

try:
    for state in map_state_list:
        years = os.listdir(f"{map_state_path}/{state}")

        for year in years:
            quaters = os.listdir(f"{map_state_path}/{state}/{year}")

            for quater in quaters:
                path = f"{map_state_path}/{state}/{year}/{quater}"

                with open(path, 'r') as f:
                    data = json.load(f)

                geo_entries = data.get('data', {}).get('data', {}).get('data', [])

                for entry in geo_entries:
                    lat = entry[0] if len(entry) > 0 else None
                    lng = entry[1] if len(entry) > 1 else None
                    metric = entry[2] if len(entry) > 2 else None
                    district = entry[3] if len(entry) > 3 else None

                    map_inscountry_data['State'].append(state)
                    map_inscountry_data['Year'].append(year)
                    map_inscountry_data['Quater'].append(int(quater.strip('.json')))
                    map_inscountry_data['District'].append(district)
                    map_inscountry_data['Metric'].append(metric)
                    map_inscountry_data['Latitude'].append(lat)
                    map_inscountry_data['Longitude'].append(lng)

except Exception as e:
    print("An error occurred during user data extraction:", e)

# Final result
map_inscountry_data
```

```python
agg_us_data = {
    'State': [], 'Year': [], 'Quater': [],
    'Brand': [], 'User_Count': [], 'User_Percentage': [],
    'Total_Registered_Users': [], 'Total_App_Opens': []
}

# Data extraction logic with error handling
try:
    for state in map_state_list:
        years = os.listdir(f"{agg_state_path}/{state}")

        for year in years:
            quaters = os.listdir(f"{agg_state_path}/{state}/{year}")

            for quater in quaters:
                path = f"{agg_state_path}/{state}/{year}/{quater}"

                # Step 1: Load JSON safely
                try:
                    with open(path, 'r') as f:
                        data = json.load(f)
                except Exception as e:
                    print(f"Failed to load JSON from {path}: {e}")
                    continue  # Skip this file

                # Step 2: Extract totals
                aggregated = data.get('data', {}).get('aggregated', {})
                total_registered = aggregated.get('registeredUsers', None)
                total_app_opens = aggregated.get('appOpens', None)

                # Step 3: Extract device usage if available
                devices = data.get('data', {}).get('usersByDevice')
                if devices and isinstance(devices, list):
                    for device in devices:
                        brand = device.get('brand', None)
                        count = device.get('count', None)
                        percentage = device.get('percentage', None)

                        # Append to final dictionary
                        agg_us_data['State'].append(state)
                        agg_us_data['Year'].append(year)
                        agg_us_data['Quater'].append(int(quater.strip('.json')))
                        agg_us_data['Brand'].append(brand)
                        agg_us_data['User_Count'].append(count)
                        agg_us_data['User_Percentage'].append(percentage)
                        agg_us_data['Total_Registered_Users'].append(total_registered)
                        agg_us_data['Total_App_Opens'].append(total_app_opens)
                else:
                    print(f"No device data found in {path}")

except Exception as e:
    print("An error occurred during user data extraction:", e)
```

```python
#-- 3. Payment categories by transaction volume
query3='''
SELECT "Category" AS Transaction_Type, SUM("Transaction_Count") AS total_count
FROM agg_transcation_data
GROUP BY Transaction_Type
ORDER BY total_count DESC;

'''

od3=pd.read_sql(query3,engine)
od3
```

```python
#-- 4. Top 5 states with declining transaction amounts over quarters
query4='''SELECT "State","Year","Quater", SUM("Transaction_Amount") AS total_amount
FROM agg_transcation_data
GROUP BY "State","Year","Quater"
ORDER BY "State","Year","Quater";
'''
od4=pd.read_sql(query4,engine)
od4
```

Business Case 2: Device Dominance and User Engagement
Goal: Understand device brand usage across regions and time.


```python
#-- 1. Registered users by device brand
query6='''SELECT "Brand" AS device_brand, SUM("Total_Registered_Users") AS total_users
FROM agg_user_data
GROUP BY device_brand
ORDER BY total_users DESC;
'''
od6=pd.read_sql(query6,engine)
od6
```

```python
#2. Regions consistently high in registrations across quarters

query22 = '''
SELECT 
    "State", 
    COUNT(DISTINCT CONCAT("Year", '-', "Quater")) AS active_quarters,
    SUM("Total_Registered_Users") AS total_registrations
FROM agg_user_data
GROUP BY "State"
HAVING COUNT(DISTINCT CONCAT("Year", '-', "Quater")) >= 4
ORDER BY total_registrations DESC;
'''

od22 = pd.read_sql(query22, engine)
od22
```

```python
#3. Regions with sudden spike in registrations with 50%

query23 = '''
WITH quarter_data AS (
    SELECT 
        "State", 
        "Year", 
        "Quater",
        SUM("Total_Registered_Users") AS users,
        ROW_NUMBER() OVER (PARTITION BY "State" ORDER BY "Year", "Quater") AS rn
    FROM agg_user_data
    GROUP BY "State", "Year", "Quater"
),
curr_prev AS (
    SELECT 
        curr."State",
        curr.users AS current_users,
        prev.users AS previous_users,
        ROUND(100.0 * (curr.users - prev.users) / NULLIF(prev.users, 0), 2) AS growth_percent
    FROM quarter_data curr
    JOIN quarter_data prev 
        ON curr."State" = prev."State" AND curr.rn = prev.rn + 1
)
SELECT * FROM curr_prev
WHERE growth_percent > 50
ORDER BY growth_percent DESC
LIMIT 10;


'''

od23 = pd.read_sql(query23, engine)
od23
```
